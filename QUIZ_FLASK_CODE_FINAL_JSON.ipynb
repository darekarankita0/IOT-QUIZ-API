```python
from flask import Flask,jsonify
import json

app = Flask(__name__)
@app.route('/', methods=['GET','POST'])
def quiz():
    q1 = """1. _________ allows us to control electronic components
    a) RETful API
    b) RESTful API
    c) HTTP
    d) MQTT"""
    q1 = {
       'Question1' : q1,
    }
    q1 = json.dumps(q1)
    q1 =jsonify(q1)

    q2 = """
    2. MQTT stands for _____________
    a) MQ Telemetry Things
    b) MQ Transport Telemetry
    c) MQ Transport Things
    d) MQ Telemetry Transport"""
    q2 = {
       'Question2' : q2,
    }
    
    q2 = json.dumps(q2)
    q2 =jsonify(q2)


    q3 = """3. MQTT is better than HTTP for sending and receiving data.
    a) True
    b) False"""
    q3 = {
       'Question3' : q3,
    }
    q3 = json.dumps(q3)
    q3 =jsonify(q3)


    q4 = """4. MQTT is _________ protocol.
    a) Machine to Machine
    b) Internet of Things
    c) Machine to Machine and Internet of Things
    d) Machine Things"""
    q4 = {
       'Question4' : q4,
    }
    
    q4 = json.dumps(q4)
    q4 =jsonify(q4)

    q5 = """5. Which protocol is lightweight?
    a) MQTT
    b) HTTP
    c) CoAP
    d) SPI"""
    q5 = {
       'Question5' : q5,
    }
    q5 = json.dumps(q5)
    q5 =jsonify(q5)
    
    questions = {q1:"a",q2:"d",q3:"a",q4:"c",q5:"a"}
    print("IOT QUIZ")
    score = 0
    for i in questions:
        #que = json.dumps(questions)
        #query=jsonify(que)
        print()
        print(i)
        answer = input("Enter the answer (a/b/c/d):")
        if answer == questions[i]:
            score = score + 1
        else:
            score = score - 1
        result=json.dumps(score)
    return jsonify(result)
           
if __name__ == '__main__':
    app.run(debug=False)
```
